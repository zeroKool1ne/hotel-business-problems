## Drop columns with high NAN-count

```python
missing_ratio = df.isna().mean().sort_values(ascending=False)
missing_ratio
```

```python
# Filter columns
# filter columns
threshold_nan = .9

cols_to_drop = missing_ratio[missing_ratio > threshold_nan].index
cols_to_drop
```
