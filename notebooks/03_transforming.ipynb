# Transformation Notebook


## Import cleaned dataset

```python
import pandas as pd

# load cleaned dataset from github
df_clean = pd.read_csv("https://raw.githubusercontent.com/zeroKool1ne/hotel-business-problems/refs/heads/main/data/processed/hotel_dataset_cleaned")
```

## Import and apply functions from HotelTransformer-Class

```python
from src.transformation import HotelTransformer  

cat_cols = ["hotel", "meal", "market_segment"]  # actual categoricals
num_cols = ["lead_time", "adr", "stays_in_week_nights"]  # actual numerics

transformer = HotelTransformer(
    cat_cols=cat_cols,
    num_cols=num_cols,
    do_scaling=True,
)

# transform dataset
df_transformed = transformer.fit_transform(df_clean)

```

## Save transformed dataset to target repo

```python
# save transformed dataset to path
pd.DataFrame(df_transformed).to_csv("../data/processed/hotel_dataset_transformed.csv", index=False)
```
